# Data Description

**Diamonds**
Four main features(also known as 4 C's) which are used to access the quality of a diamond, are listed down below.

- Carat Weight
- Cut
- Clarity
- Color
- Carat Weight

Carat is a metric that is used to measure the weight of a diamond. One carat is equivalent to 200mg. Diamond prices increase with diamond carat weight, which means bigger the diamond higher the price. If two diamonds weights are equal, then other features are used to determine the price.

**Clarity**
Diamonds are generated from sheer pressure and heat below the ground. Therefore, there will be some inclusion inside a diamond i.e., a mark or line pattern inside a diamond. Also, there will be a mark or line in the outer layer of a diamond, which is called blemishes. Based on the amount of inclusion and blemishes, the clarity of a diamond is categorized such as FL, IF, VSS1, VSS2, VS1, VS2, SI1, SI2, I1, I2, I3. The categories mentioned above are ordered in descending order by the amount of presence of inclusion and blemishes.

**Cut** : This feature is an important thing to notice in a diamond as it measures three crucial things, such as :

**Brilliance**: It means the brightness of a diamond by the reflection of white lights inside and outside of a diamond.

**Fire**: It means Scattering of white light into all the colors of the rainbow.
**Scintillation**: the amount of sparkle produced and the pattern of light and dark areas caused by reflection within a diamond.

The goal is to cut a diamond within an appropriate size shape, and angle such that the light entering the diamond should reflect and leave from the top surface. The other Features of a Cut are the Depth Percentage and Table percentage.

**Depth %**
Depth is the distance from a top surface i.e., table to a culet. The depth percentage is calculated by dividing the diamond depth by the overall width of a diamond. Lower the depth percentage the bigger the diamond looks from the below i.e., pavilion.


**Table %**
The table is the topmost surface of a diamond and also the most significant facet of the round diamond. An appropriate width of a table will allow the light to enter and reflect on the appropriate direction .if not most of the light will scatter off in different directions. The table percentage is calculated by dividing the table width by overall diamond width. The following five categories measure the quality of Cut.
- **Excellent**
- **Very Good**
- **Good**
- **Fair**
- **Poor**

**Color**
Color measurement in diamond measures lacks color. If the diamond color is like a drop of water that is colorless, it will have a high value. As then only it can scatter the light without observing. However, there are some diamonds that are in different colors will have higher prices, but here in the dataset, we are going to only look at the colorless diamonds.

The color scale is categorized from D to Z letters and ordered in ascending by the amount of presence of color in a diamond. From the K onwards to till Z, we can see a yellowish color present.

D ,E,F - Colorless G,H,I,J - Near colorlessness K, L, M - Faint color N-R: Very Light Color S-Z: light color

Length,Width, and Depth
The dimension of a diamond is measured in millimeters. Moreover, the shape of a diamond is determined by the Length to width ratio. For instance, to determine the roundness of a diamond, we need to check the L/W ratio, If the ratio is between 1 and 1.05, it is a round diamond, and an oval shape diamond L/W ratio can be around 1.50 or less.

# Part 4 - Inferences

### Proportion of Diamond by clarity

```python
# Count the number of diamonds with each clarity grade
clarity_counts = Diamond_data['clarity'].value_counts()

# Calculate the proportion of diamonds with a clarity of SI1, SI2, or VS2
total_diamonds = len(Diamond_data)
proportion_of_diamonds = (clarity_counts['SI1'] + clarity_counts['SI2'] + clarity_counts['VS2']) / total_diamonds

# Print the proportion of diamonds
print(proportion_of_diamonds)
```

### Diamond with the lenght less than   of the Mean

```python
# Calculate the length of each diamond in millimeters
Diamond_data['length_mm'] = np.sqrt((Diamond_data['x']**2) + (Diamond_data['y']**2))

# Calculate the mean diamond length
mean_length = Diamond_data['length_mm'].mean()

# Calculate the length threshold
three_quarters_mean_length = mean_length * 0.75

# Filter the dataset to only include diamonds with a length less than the threshold
filtered_df = Diamond_data[Diamond_data['length_mm'] < three_quarters_mean_length]

# Count the number of diamonds in the filtered dataset
num_diamonds = len(filtered_df)

# Print the number of diamonds
print(num_diamonds)
```

### Diamonds with Carat Value Greater Than the Median

```python
# Calculate the median carat value
median_carat = Diamond_data['carat'].median()

# Filter the dataset to only include diamonds with a carat value greater than the median
filtered_df = Diamond_data[Diamond_data['carat'] > median_carat]

# Count the number of diamonds in the filtered dataset
num_diamonds = len(filtered_df)

# Print the number of diamonds
print(num_diamonds)
```

### Diamonds with 'Fair' or 'Premium' Cut

```python
# Filter the dataset to only include diamonds with a cut of Fair or Premium
filtered_df = Diamond_data[(Diamond_data['cut']=='Fair') | (Diamond_data['cut']=='Premium')]

# Count the number of diamonds in the filtered dataset
num_diamonds = len(filtered_df)

# Print the number of diamonds
print(num_diamonds)
```

### Lowest Price per Carat

```python
# Calculate the price per carat for each diamond
Diamond_data['price_per_carat'] = Diamond_data['price'] / Diamond_data['carat']

# Find the index of the diamond with the lowest price per carat
min_price_per_carat_index = Diamond_data['price_per_carat'].idxmin()

# Get the price per carat for the diamond with the lowest price per carat
min_price_per_carat = Diamond_data.loc[min_price_per_carat_index, 'price_per_carat']

# Print the dataframe row index and the price per carat for the diamond with the lowest price per carat
print("Minimum Price per carat index : " ,min_price_per_carat_index)
print("Minimum  price per carat : " ,min_price_per_carat)
```

### Boxplots of Carat Metric by Clarity

### Scatter Plot of Carat vs. Depth

```python
# Create a scatter plot using Matplotlib
plt.figure(figsize=(10, 6))
plt.scatter(Diamond_data['carat'], Diamond_data['depth'], s=10, marker='o', alpha=0.5)
plt.xlabel('Carat')
plt.ylabel('Depth')
plt.title('Carat vs. Depth (Matplotlib)')
plt.grid(True)
plt.show()
```

```python
# Create a scatter plot using Seaborn
sns.jointplot(x='carat', y='depth', data=Diamond_data, size=10, alpha=0.5)
plt.title('Carat vs. Depth (Seaborn)')
plt.grid(True)
plt.show()
```

Both the Matplotlib and Seaborn scatter plots show that there is a positive correlation between carat and depth. This means that, in general, larger diamonds tend to have a greater depth.

The Seaborn scatter plot also shows that the spread of the data points increases with carat size. This suggests that there is more variation in the depth of larger diamonds.
